```python
# Day 16: 30 Days of Python - Python Date and Time

from datetime import datetime, date, timedelta

# 1️⃣ Get the current day, month, year, hour, minute and timestamp
now = datetime.now()
print("Current datetime:", now)

print("Year:", now.year)
print("Month:", now.month)
print("Day:", now.day)
print("Hour:", now.hour)
print("Minute:", now.minute)
print("Timestamp:", now.timestamp())
```

```python

# 2️⃣ Format the current date using "%m/%d/%Y, %H:%M:%S"
formatted_date = now.strftime("%m/%d/%Y, %H:%M:%S")
print("Formatted date:", formatted_date)
```

```python
# 3️⃣ Convert string "5 December, 2019" to datetime object
date_string = "5 December, 2019"
converted_date = datetime.strptime(date_string, "%d %B, %Y")
print("Converted Date:", converted_date)
```

```python

# 4️⃣ Calculate time difference between now and new year
new_year = datetime(now.year + 1, 1, 1)
time_to_new_year = new_year - now
print("Time remaining to New Year:", time_to_new_year)
print("Days remaining:", time_to_new_year.days)
```

```python
# 5️⃣ Calculate the time difference between 1 January 1970 and now
past = datetime(1970, 1, 1)
time_difference = now - past
print("Time difference since 1970:", time_difference)
print("Total days since 1970:", time_difference.days)
print("Total seconds since 1970:", time_difference.total_seconds())
```
