```python
# Day 10: 30 Days of Python - Loops

# --------------------------------------
# Exercises: Level 1
# --------------------------------------

# 1Ô∏è Iterate 0 to 10 using for loop
for i in range(11):
    print(i, end=" ")
print("\n")

# 1 Iterate 0 to 10 using while loop
i = 0
while i <= 10:
    print(i, end=" ")
    i += 1
print("\n")
```

```python
# 2Ô∏è Iterate 10 to 0 using for loop
for i in range(10, -1, -1):
    print(i, end=" ")
print("\n")

# 2Ô∏è Iterate 10 to 0 using while loop
i = 10
while i >= 0:
    print(i, end=" ")
    i -= 1
print("\n")
```

```python
# 3Ô∏è Print a triangle pattern
for i in range(1, 8):
    print("#" * i)
print("\n")
```

```python
# 4Ô∏è Nested loops for grid pattern
for i in range(8):
    for j in range(8):
        print("#", end=" ")
    print()
print("\n")
```

```python
# 5Ô∏è Multiplication pattern
for i in range(11):
    print(f"{i} x {i} = {i * i}")
print("\n")
```

```python
# 6Ô∏è Iterate through a list and print items
tech_stack = ['Python', 'Numpy', 'Pandas', 'Django', 'Flask']
for item in tech_stack:
    print(item)
print("\n")
```

```python
# 7Ô∏è Print only even numbers (0‚Äì100)
for i in range(101):
    if i % 2 == 0:
        print(i, end=" ")
print("\n")
```

```python
# 8Ô∏è Print only odd numbers (0‚Äì100)
for i in range(101):
    if i % 2 != 0:
        print(i, end=" ")
print("\n")
```

```python

# 9Ô∏è‚É£ Sum of all numbers (0‚Äì100)
total_sum = 0
for i in range(101):
    total_sum += i
print(f"The sum of all numbers is {total_sum}")
print()
```

```python
# üîü Sum of all evens and odds
even_sum = 0
odd_sum = 0
for i in range(101):
    if i % 2 == 0:
        even_sum += i
    else:
        odd_sum += i
print(f"The sum of all evens is {even_sum}. And the sum of all odds is {odd_sum}.")
print()
```

```python
# 2Ô∏è Reverse a fruit list using a loop
fruits = ['banana', 'orange', 'mango', 'lemon']
reversed_fruits = []
for i in range(len(fruits) - 1, -1, -1):
    reversed_fruits.append(fruits[i])
print("Reversed fruit list:", reversed_fruits)
print()
```

```python
# 3Ô∏è Analyze countries_data.py file
from data.countries_data import countries_data  # Should contain country info dictionaries

# Total number of languages
all_languages = []
for country in countries_data:
    all_languages.extend(country['languages'])
unique_languages = set(all_languages)
print("Total number of languages:", len(unique_languages))
print()
```

```python
# 10 Ten most spoken languages
from collections import Counter
language_counter = Counter(all_languages)
most_spoken = language_counter.most_common(10)
print("Ten most spoken languages:")
for lang, count in most_spoken:
    print(f"{lang}: {count}")
print()
```

```python
# 10 most populated countries
sorted_by_population = sorted(countries_data, key=lambda x: x['population'], reverse=True)
print("Ten most populated countries:")
for country in sorted_by_population[:10]:
    print(f"{country['name']}: {country['population']}")
```
