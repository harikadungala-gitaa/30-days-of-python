```python
# Day 17: 30 Days of Python - Unpacking Exercise

# Given list
names = ['Finland', 'Sweden', 'Norway', 'Denmark', 'Iceland', 'Estonia', 'Russia']

# Unpacking
*nordic_countries, es, ru = names

# Print results
print("Nordic Countries:", nordic_countries)
print("Estonia:", es)
print("Russia:", ru)
```
