```python
# Day 15: 30 Days of Python - Python Error Types

# ⚙️ SyntaxError
# print 'hello world'  # Invalid Syntax
print('hello world')   #  Correct

# ⚙️ NameError
try:
    print(age)
except NameError as e:
    print("NameError:", e)
age = 25
print("Fixed:", age)

# ⚙️ IndexError
numbers = [1, 2, 3, 4, 5]
try:
    print(numbers[5])
except IndexError as e:
    print("IndexError:", e)

# ⚙️ ModuleNotFoundError
try:
    import maths
except ModuleNotFoundError as e:
    print("ModuleNotFoundError:", e)
import math
print("PI from math module:", math.pi)

# ⚙️ AttributeError
try:
    print(math.PI)
except AttributeError as e:
    print("AttributeError:", e)
print("Correct value:", math.pi)

# ⚙️ KeyError
user = {'name': 'Asab', 'age': 250, 'country': 'Finland'}
try:
    print(user['county'])
except KeyError as e:
    print("KeyError:", e)
print("Fixed:", user['country'])

# ⚙️ TypeError
try:
    print(4 + '3')
except TypeError as e:
    print("TypeError:", e)
print("Fixed:", 4 + int('3'))

# ⚙️ ImportError
try:
    from math import power
except ImportError as e:
    print("ImportError:", e)
from math import pow
print("Fixed:", pow(2, 3))

# ⚙️ ValueError
try:
    print(int('12a'))
except ValueError as e:
    print("ValueError:", e)

# ⚙️ ZeroDivisionError
try:
    print(1 / 0)
except ZeroDivisionError as e:
    print("ZeroDivisionError:", e)
```
