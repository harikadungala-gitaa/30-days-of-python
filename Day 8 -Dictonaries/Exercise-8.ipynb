```python
# Day 8: 30 Days of Python Programming
# Level 1 – Dictionaries

# 1. Create an empty dictionary called dog
dog = {}
print("Empty dog dictionary:", dog)
```

```python
# 2. Add name, color, breed, legs, age to the dog dictionary
dog['name'] = 'Buddy'
dog['color'] = 'Brown'
dog['breed'] = 'Golden Retriever'
dog['legs'] = 4
dog['age'] = 5
print("\nDog dictionary:", dog)
```

```python
# 3. Create a student dictionary with the given keys
student = {
    'first_name': 'Harika',
    'last_name': 'Dungala',
    'gender': 'Female',
    'age': 21,
    'marital_status': 'Single',
    'skills': ['Python', 'SQL'],
    'country': 'India',
    'city': 'Hyderabad',
    'address': {
        'street': 'MG Road',
        'zipcode': '500081'
    }
}
print("\nStudent dictionary:", student)
```

```python
# 4. Get the length of the student dictionary
print("\nLength of student dictionary:", len(student))
```

```python
# 5. Get the value of 'skills' and check its data type
skills_value = student['skills']
print("\nSkills:", skills_value)
print("Data type of skills:", type(skills_value))
```

```python
# 6. Modify the skills values by adding one or two more skills
student['skills'].extend(['JavaScript', 'HTML'])
print("Updated skills:", student['skills'])
```

```python
# 7. Get the dictionary keys as a list
keys_list = list(student.keys())
print("\nDictionary keys:", keys_list)
```

```python
# 8. Get the dictionary values as a list
values_list = list(student.values())
print("Dictionary values:", values_list)
```

```python
# 9. Change the dictionary to a list of tuples using items()
items_list = list(student.items())
print("\nDictionary as list of tuples:", items_list)
```

```python
# 10. Delete one of the items in the dictionary
student.pop('marital_status')
print("\nAfter deleting 'marital_status':", student)
```

```python
# 11. Delete one of the dictionaries (let’s delete dog)
del dog
print("\nDog dictionary deleted successfully!")
```
