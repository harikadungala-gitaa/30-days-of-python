```python
# Day 18: 30 Days of Python - Regular Expressions

import re
from collections import Counter

# -------------------------------
# ðŸ§© Exercise Level 1
# -------------------------------

paragraph = '''I love teaching. If you do not love teaching what else can you love. 
I love Python if you do not love something which can give you all the capabilities 
to develop an application what else can you love.'''

# Clean punctuation and lowercase everything
words = re.findall(r'\b\w+\b', paragraph.lower())
word_counts = Counter(words)

# Sort by frequency (descending)
most_common = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)
print("Most frequent words:")
print(most_common)

# -------------------------------
# ðŸ§© Distance between two furthest points
# -------------------------------

text = "The position of some particles on the horizontal x-axis are -12, -4, -3, -1 in the negative direction, 0 at origin, 4 and 8 in the positive direction."
points = re.findall(r'-?\d+', text)  # Extract all integers (with possible minus)
points = [int(p) for p in points]
points.sort()

print("\nExtracted points:", points)
distance = max(points) - min(points)
print("Distance between furthest points:", distance)
```

```python

#  Exercise Level 2


def is_valid_variable(var):
    # A valid variable name:
    # - starts with letter or underscore
    # - followed by letters, numbers, or underscores
    return bool(re.match(r'^[A-Za-z_]\w*$', var))

# Test examples
print(is_valid_variable('first_name'))   # True
print(is_valid_variable('first-name'))   # False
print(is_valid_variable('1first_name'))  # False
print(is_valid_variable('firstname'))    # True
```

```python

#  Exercise Level 3


sentence = '''%I $am@% a %tea@cher%, &and& I lo%#ve %tea@ching%;.
There $is nothing; &as& mo@re rewarding as educa@ting &and& @emp%o@wering peo@ple.
;I found tea@ching m%o@re interesting tha@n any other %jo@bs.
%Do@es thi%s mo@tivate yo@u to be a tea@cher!?'''

def clean_text(text):
    # Remove all non-alphabetic and non-space characters
    cleaned = re.sub(r'[^A-Za-z\s]', '', text)
    return cleaned

def most_frequent_words(text):
    words = re.findall(r'\b\w+\b', text)
    counts = Counter(words)
    return counts.most_common(3)

cleaned_text = clean_text(sentence)
print("\nCleaned text:")
print(cleaned_text)

print("\nTop 3 frequent words:")
print(most_frequent_words(cleaned_text))
```
