```python
# 1Ô∏è‚É£ Declare a function add_two_numbers. It takes two parameters and returns a sum.

def add_two_numbers(a, b):
    return a + b
```

```python
# 2Ô∏è‚É£ Area of a circle (area = œÄr¬≤)

import math
def area_of_circle(r):
    return math.pi * r ** 2
```

```python
# 3Ô∏è‚É£ Function add_all_nums which takes arbitrary number of arguments and sums them.

def add_all_nums(*args):
    if all(isinstance(i, (int, float)) for i in args):
        return sum(args)
    else:
        return "All arguments must be numbers."
```

```python
# 4Ô∏è‚É£ Convert Celsius to Fahrenheit

def convert_celsius_to_fahrenheit(c):
    return (c * 9/5) + 32
```

```python
# 5Ô∏è‚É£ Function check_season which returns the season based on month.

def check_season(month):
    month = month.capitalize()
    if month in ["September", "October", "November"]:
        return "Autumn"
    elif month in ["December", "January", "February"]:
        return "Winter"
    elif month in ["March", "April", "May"]:
        return "Spring"
    elif month in ["June", "July", "August"]:
        return "Summer"
    else:
        return "Invalid month"
```

```python
# 6Ô∏è‚É£ Function calculate_slope

def calculate_slope(x1, y1, x2, y2):
    if x2 == x1:
        return "Undefined slope (vertical line)"
    return (y2 - y1) / (x2 - x1)
```

```python
# 7Ô∏è‚É£ Function solve_quadratic_eqn

def solve_quadratic_eqn(a, b, c):
    d = b ** 2 - 4 * a * c
    if d > 0:
        root1 = (-b + math.sqrt(d)) / (2 * a)
        root2 = (-b - math.sqrt(d)) / (2 * a)
        return (root1, root2)
    elif d == 0:
        return (-b / (2 * a),)
    else:
        return "No real roots"
```

```python
# 8Ô∏è‚É£ Function print_list

def print_list(lst):
    for item in lst:
        print(item)
```

```python
# 9Ô∏è‚É£ Function reverse_list

def reverse_list(lst):
    reversed_list = []
    for i in range(len(lst) - 1, -1, -1):
        reversed_list.append(lst[i])
    return reversed_list
```

```python
# üîü Function capitalize_list_items

def capitalize_list_items(lst):
    return [item.capitalize() for item in lst]
```

```python
# 1Ô∏è‚É£1Ô∏è‚É£ Function add_item

def add_item(lst, item):
    lst.append(item)
    return lst
```

```python
# 1Ô∏è‚É£2Ô∏è‚É£ Function remove_item

def remove_item(lst, item):
    if item in lst:
        lst.remove(item)
    return lst
```

```python
# 1Ô∏è‚É£3Ô∏è‚É£ Function sum_of_numbers

def sum_of_numbers(n):
    return sum(range(1, n + 1))
```

```python
# 1Ô∏è‚É£4Ô∏è‚É£ Function sum_of_odds

def sum_of_odds(n):
    return sum(i for i in range(1, n + 1) if i % 2 != 0)
```

```python
# 1Ô∏è‚É£5Ô∏è‚É£ Function sum_of_evens

def sum_of_evens(n):
    return sum(i for i in range(1, n + 1) if i % 2 == 0)
```

```python
# 1Ô∏è‚É£ Function evens_and_odds

def evens_and_odds(n):
    evens = len([i for i in range(n + 1) if i % 2 == 0])
    odds = len([i for i in range(n + 1) if i % 2 != 0])
    return f"The number of odds are {odds}. The number of evens are {evens}."
```

```python
# 2Ô∏è‚É£ Function factorial

def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
```

```python
# 3Ô∏è‚É£ Function is_empty

def is_empty(item):
    return not bool(item)
```

```python
# 4Ô∏è‚É£ Statistical functions

from statistics import mean, median, mode, variance, stdev

def calculate_mean(lst):
    return mean(lst)

def calculate_median(lst):
    return median(lst)

def calculate_mode(lst):
    return mode(lst)

def calculate_range(lst):
    return max(lst) - min(lst)

def calculate_variance(lst):
    return variance(lst)

def calculate_std(lst):
    return stdev(lst)
```

```python
# 1Ô∏è‚É£ Function is_prime

import math
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True
```

```python
# 2Ô∏è‚É£ Function all_unique

def all_unique(lst):
    return len(lst) == len(set(lst))
```

```python
# 3Ô∏è‚É£ Function same_data_type

def same_data_type(lst):
    return all(type(i) == type(lst[0]) for i in lst)
```

```python
# 4Ô∏è‚É£ Function is_valid_variable

def is_valid_variable(var):
    return var.isidentifier()
```

```python
# 5Ô∏è‚É£ Function most_spoken_languages

from collections import Counter
from data.countries_data import countries_data

def most_spoken_languages(data=countries_data, top_n=10):
    all_languages = []
    for country in data:
        all_languages.extend(country["languages"])
    lang_count = Counter(all_languages)
    return lang_count.most_common(top_n)
```

```python
# 6Ô∏è‚É£ Function most_populated_countries

def most_populated_countries(data=countries_data, top_n=10):
    sorted_countries = sorted(data, key=lambda x: x["population"], reverse=True)
    return [(country["name"], country["population"]) for country in sorted_countries[:top_n]]
```
