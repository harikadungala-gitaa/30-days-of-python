```python
# Day 12: 30 Days of Python - Modules (Random, String)

import random
import string

# --------------------------------------
#  Exercises: Level 1
# --------------------------------------

# 1️ random_user_id – generates a 6-character ID
def random_user_id():
    characters = string.ascii_letters + string.digits
    user_id = ''.join(random.choice(characters) for _ in range(6))
    return user_id
```

```python
# 2️⃣ user_id_gen_by_user – asks for length and number of IDs
def user_id_gen_by_user():
    num_chars = int(input("Enter number of characters per ID: "))
    num_ids = int(input("Enter number of IDs to generate: "))
    characters = string.ascii_letters + string.digits
    for _ in range(num_ids):
        user_id = ''.join(random.choice(characters) for _ in range(num_chars))
        print(user_id)
```

```python
# 3️⃣ rgb_color_gen – generates RGB string
def rgb_color_gen():
    r = random.randint(0, 255)
    g = random.randint(0, 255)
    b = random.randint(0, 255)
    return f"rgb({r},{g},{b})"
```

```python
# 4️⃣ list_of_hexa_colors – returns list of hex colors
def list_of_hexa_colors(num):
    hex_colors = []
    for _ in range(num):
        hex_color = '#' + ''.join(random.choice('0123456789abcdef') for _ in range(6))
        hex_colors.append(hex_color)
    return hex_colors
```

```python
# 5️⃣ list_of_rgb_colors – returns list of RGB colors
def list_of_rgb_colors(num):
    rgb_colors = []
    for _ in range(num):
        r = random.randint(0, 255)
        g = random.randint(0, 255)
        b = random.randint(0, 255)
        rgb_colors.append(f"rgb({r},{g},{b})")
    return rgb_colors
```

```python
# 6️⃣ generate_colors – can create RGB or HEX lists
def generate_colors(color_type, num):
    if color_type.lower() == 'hexa':
        return list_of_hexa_colors(num)
    elif color_type.lower() == 'rgb':
        return list_of_rgb_colors(num)
    else:
        return "Invalid color type"
```

```python
# 7️⃣ shuffle_list – returns a shuffled list
def shuffle_list(lst):
    random.shuffle(lst)
    return lst
```

```python
# 8️⃣ unique_random_numbers – 7 unique random numbers 0–9
def unique_random_numbers():
    numbers = random.sample(range(10), 7)
    return numbers
```
