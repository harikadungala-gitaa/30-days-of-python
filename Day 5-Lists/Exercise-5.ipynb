```python
# Level 2 Exercises

# 1. Given list of ages
ages = [19, 22, 19, 24, 20, 25, 26, 24, 25, 24]
print("Original ages:", ages)
```

```python
# 2. Sort the list and find the min and max age
ages.sort()
print("Sorted ages:", ages)
min_age = min(ages)
max_age = max(ages)
print("Min age:", min_age)
print("Max age:", max_age)
```

```python
# 3. Add the min age and max age again to the list
ages.append(min_age)
ages.append(max_age)
print("Ages after adding min and max again:", ages)
```

```python
# 4. Find the median age
n = len(ages)
ages.sort()
if n % 2 == 0:
    median_age = (ages[n//2 - 1] + ages[n//2]) / 2
else:
    median_age = ages[n//2]
print("Median age:", median_age)
```

```python

# 5. Find the average age
average_age = sum(ages) / len(ages)
print("Average age:", average_age)
```

```python
# 6. Find the range of the ages
age_range = max_age - min_age
print("Age range:", age_range)
```

```python
# 7. Compare (min - average) and (max - average)
diff_min_avg = abs(min_age - average_age)
diff_max_avg = abs(max_age - average_age)
print("Difference (min - average):", diff_min_avg)
print("Difference (max - average):", diff_max_avg)
```

```python
# 8. Find the middle country(ies)
countries = ['China', 'Russia', 'USA', 'Finland', 'Sweden', 'Norway', 'Denmark']
print("\nCountries:", countries)
mid_index = len(countries) // 2
if len(countries) % 2 == 0:
    middle_countries = countries[mid_index-1:mid_index+1]
else:
    middle_countries = [countries[mid_index]]
print("Middle country(ies):", middle_countries)
```

```python
# 9. Divide countries into two equal halves
if len(countries) % 2 == 0:
    first_half = countries[:len(countries)//2]
    second_half = countries[len(countries)//2:]
else:
    first_half = countries[:len(countries)//2 + 1]
    second_half = countries[len(countries)//2 + 1:]
print("First half:", first_half)
print("Second half:", second_half)
```

```python
# 10. Unpack the first three countries and the rest as scandic countries
china, russia, usa, *scandic = countries
print("\nUnpacked countries:")
print("China:", china)
print("Russia:", russia)
print("USA:", usa)
print("Scandic countries:", scandic)
```
