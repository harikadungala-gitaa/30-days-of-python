```python
# Day 6: 30 Days of Python Programming
# Level 1 Exercises - Tuples

# 1. Create an empty tuple
empty_tuple = ()
print("Empty tuple:", empty_tuple)
```

```python
# 2. Create a tuple containing names of your sisters and brothers
brothers = ('Ravi', 'Kiran', 'Ajay')
sisters = ('Priya', 'Harika', 'Sita')
print("Brothers:", brothers)
print("Sisters:", sisters)
```

```python
# 3. Join brothers and sisters tuples and assign it to siblings
siblings = brothers + sisters
print("Siblings:", siblings)
```

```python
# 4. How many siblings do you have?
print("Number of siblings:", len(siblings))
```

```python
# 5. Modify the siblings tuple and add the name of your father and mother to create family_members
# (Tuples are immutable, so we create a new tuple)
family_members = siblings + ('Father', 'Mother')
print("Family members:", family_members)
```

LEVEL-2

```python
# Day 6: 30 Days of Python Programming
# Level 2 Exercises - Tuples

# 1. Unpack siblings and parents from family_members
family_members = ('Ravi', 'Kiran', 'Ajay', 'Priya', 'Harika', 'Sita', 'Father', 'Mother')
siblings = family_members[:-2]
parents = family_members[-2:]
print("Siblings:", siblings)
print("Parents:", parents)
```

```python

# 2. Create fruits, vegetables, and animal products tuples
fruits = ('apple', 'banana', 'mango', 'orange')
vegetables = ('carrot', 'potato', 'spinach', 'brinjal')
animal_products = ('milk', 'cheese', 'eggs', 'meat')
```

```python
# 3. Join the three tuples and assign it to a variable called food_stuff_tp
food_stuff_tp = fruits + vegetables + animal_products
print("\nFood stuff tuple:", food_stuff_tp)
```

```python

# 4. Change the food_stuff_tp tuple to a food_stuff_lt list
food_stuff_lt = list(food_stuff_tp)
print("Food stuff list:", food_stuff_lt)
```

```python
# 5. Slice out the middle item(s) from the food_stuff_tp or food_stuff_lt
mid_index = len(food_stuff_lt) // 2
if len(food_stuff_lt) % 2 == 0:
    middle_items = food_stuff_lt[mid_index - 1:mid_index + 1]
else:
    middle_items = [food_stuff_lt[mid_index]]
print("\nMiddle item(s):", middle_items)
```

```python
# 6. Slice out the first three items and the last three items from food_stuff_lt list
first_three = food_stuff_lt[:3]
last_three = food_stuff_lt[-3:]
print("First three items:", first_three)
print("Last three items:", last_three)
```

```python
# 7. Delete the food_stuff_tp tuple completely
del food_stuff_tp
print("\nfood_stuff_tp tuple deleted successfully!")
```
