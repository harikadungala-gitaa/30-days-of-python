```python
# Declaring variables
first_name = 'Harika'
last_name = 'Dungala'
full_name = first_name + ' ' + last_name
country = 'India'
city = 'Hyderabad'
age = 21
year = 2025
is_married = False
is_true = True
is_light_on = True

# Declaring multiple variables in one line
first_name, last_name, country, age, is_married = 'Harika', 'Dungala', 'India', 21, False
```

```python
#1. Check the data type of all your variables
print("=== Data Types ===")
print('first_name:', type(first_name))
print('last_name:', type(last_name))
print('full_name:', type(full_name))
print('country:', type(country))
print('city:', type(city))
print('age:', type(age))
print('year:', type(year))
print('is_married:', type(is_married))
print('is_true:', type(is_true))
print('is_light_on:', type(is_light_on))
```

```python
# 2. Using len() to find the length of your first name
print("\n=== Length Comparison ===")
print('Length of first name:', len(first_name))
print('Length of last name:', len(last_name))
```

```python
# 3. Compare length of first and last name
if len(first_name) > len(last_name):
    print('Your first name is longer than your last name.')
elif len(first_name) < len(last_name):
    print('Your last name is longer than your first name.')
else:
    print('Your first name and last name are of equal length.')
```

```python
# 4. Declare numbers
num_one = 5
num_two = 4

# 5â€“10. Arithmetic operations
total = num_one + num_two
diff = num_one - num_two
product = num_one * num_two
division = num_one / num_two
remainder = num_two % num_one
exp = num_one ** num_two
floor_division = num_one // num_two

print("\n=== Arithmetic Operations ===")
print('Total:', total)
print('Difference:', diff)
print('Product:', product)
print('Division:', division)
print('Remainder:', remainder)
print('Exponent:', exp)
print('Floor Division:', floor_division)
```

```python
# 11. Circle calculations
radius = 30
pi = 3.14159
area_of_circle = pi * radius ** 2
circum_of_circle = 2 * pi * radius
```

```python

print("\n=== Circle Calculations ===")
print('Radius:', radius)
print('Area of circle:', area_of_circle)
print('Circumference of circle:', circum_of_circle)
```

```python

# 12. Take radius as user input and calculate area
radius_input = float(input("\nEnter the radius of a circle: "))
area_from_input = pi * radius_input ** 2
print('Area from user input:', area_from_input)
```
