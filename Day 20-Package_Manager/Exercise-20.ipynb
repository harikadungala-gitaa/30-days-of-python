```python
#1
import requests
import re
from collections import Counter

url = 'http://www.gutenberg.org/files/1112/1112.txt'
resp = requests.get(url)
text = resp.text.lower()
words = re.findall(r"\b[a-z']+\b", text)
word_counts = Counter(words)
top10 = word_counts.most_common(10)
print("Top 10 words:", top10)
```

```python
import requests
import statistics

resp = requests.get("https://api.thecatapi.com/v1/breeds")
breeds = resp.json()

weights = []
lifespans = []
country_breed_freq = {}

for b in breeds:
    if not isinstance(b, dict):
        continue  # skip if not a dictionary

    # weight
    w = b.get('weight', {}).get('metric', '')
    if ' - ' in w:
        parts = w.split(' - ')
        try:
            avg_w = (float(parts[0]) + float(parts[1])) / 2
            weights.append(avg_w)
        except ValueError:
            pass

    # lifespan
    l = b.get('life_span', '')
    if ' - ' in l:
        parts = l.split(' - ')
        try:
            avg_l = (float(parts[0]) + float(parts[1].split()[0])) / 2
            lifespans.append(avg_l)
        except ValueError:
            pass

    # country frequency
    country = b.get('origin', 'Unknown')
    country_breed_freq[country] = country_breed_freq.get(country, 0) + 1

print("\nðŸ± Weight stats (kg):")
print("Min:", min(weights), "Max:", max(weights),
      "Mean:", statistics.mean(weights),
      "Median:", statistics.median(weights),
      "StdDev:", statistics.pstdev(weights))

print("\nðŸ¾ Lifespan stats (years):")
print("Min:", min(lifespans), "Max:", max(lifespans),
      "Mean:", statistics.mean(lifespans),
      "Median:", statistics.median(lifespans),
      "StdDev:", statistics.pstdev(lifespans))

print("\nðŸŒ Countryâ€“Breed frequency:")
for k, v in country_breed_freq.items():
    print(f"{k}: {v}")
```

```python
import requests
from collections import Counter

# Fetch countries data
resp = requests.get("https://restcountries.com/v3.1/all")
countries = resp.json()

# 10 most populated countries
pop_sorted = sorted(
    [c for c in countries if isinstance(c, dict) and 'population' in c],
    key=lambda c: c.get('population', 0),
    reverse=True
)

top10_pop = []
for c in pop_sorted[:10]:
    name = c.get('name', {}).get('common', 'Unknown')
    pop = c.get('population', 0)
    top10_pop.append({'country': name, 'population': pop})

# 10 most spoken languages
langs = []
for c in countries:
    if isinstance(c, dict) and 'languages' in c:
        langs.extend(list(c['languages'].values()))
lang_counts = Counter(langs).most_common(10)

# Total number of unique languages
total_languages = len(set(langs))

print("Top 10 most populated countries:\n", top10_pop)
print("\nTop 10 most spoken languages:\n", lang_counts)
print("\nTotal number of languages:", total_languages)
```
