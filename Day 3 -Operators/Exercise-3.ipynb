```python
# 4. Calculate area of a triangle
base = float(input("\nEnter base: "))
height_t = float(input("Enter height: "))
area_of_triangle = 0.5 * base * height_t
print("The area of the triangle is", area_of_triangle)
```

```python
# 5. Calculate perimeter of a triangle
a = float(input("\nEnter side a: "))
b = float(input("Enter side b: "))
c = float(input("Enter side c: "))
perimeter = a + b + c
print("The perimeter of the triangle is", perimeter)
```

```python
# 6. Rectangle: area and perimeter
length = float(input("\nEnter length of rectangle: "))
width = float(input("Enter width of rectangle: "))
area_rectangle = length * width
perimeter_rectangle = 2 * (length + width)
print("Area of rectangle:", area_rectangle)
print("Perimeter of rectangle:", perimeter_rectangle)
```

```python
# 7. Circle: area and circumference
radius = float(input("\nEnter radius of circle: "))
pi = 3.14
area_circle = pi * radius ** 2
circum_circle = 2 * pi * radius
print("Area of circle:", area_circle)
print("Circumference of circle:", circum_circle)
```

```python
# 8. Slope, x-intercept, and y-intercept of y = 2x - 2
# y = mx + b -> slope = 2, y-intercept = -2, x-intercept -> 0 = 2x - 2 â†’ x = 1
slope1 = 2
y_intercept = -2
x_intercept = 1
print("\nEquation: y = 2x - 2")
print("Slope:", slope1)
print("x-intercept:", x_intercept)
print("y-intercept:", y_intercept)
```

```python
# 9. Slope and distance between (2, 2) and (6, 10)
x1, y1 = 2, 2
x2, y2 = 6, 10
slope2 = (y2 - y1) / (x2 - x1)
import math
distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
print("\nBetween points (2,2) and (6,10):")
print("Slope:", slope2)
print("Euclidean distance:", distance)
```

```python
# 10. Compare slopes
print("\nComparing slopes:")
print("Are the slopes equal?", slope1 == slope2)
```

```python
# 20. Check if int('9.8') == 10 (this will cause error, so handle it safely)
try:
    result = int('9.8')
except ValueError:
    result = "Error (cannot convert '9.8' to int)"
print("Is int('9.8') == 10?", result == 10)
```
