```python
# Day 4: 30 Days of Python Programming

# 1. Concatenate strings
part1, part2, part3, part4 = 'Thirty', 'Days', 'Of', 'Python'
sentence1 = part1 + ' ' + part2 + ' ' + part3 + ' ' + part4
print(sentence1)
```

```python
# 2. Concatenate 'Coding', 'For', 'All'
sentence2 = 'Coding' + ' ' + 'For' + ' ' + 'All'
print(sentence2)
```

```python
# 3. Declare variable
company = "Coding For All"
```

```python
# 4. Print company
print(company)
```

```python
# 5. Print length
print("Length of company:", len(company))
```

```python
# 6. Uppercase
print("Uppercase:", company.upper())
```

```python
# 7. Lowercase
print("Lowercase:", company.lower())
```

```python
# 8. capitalize(), title(), swapcase()
print("Capitalize:", company.capitalize())
print("Title:", company.title())
print("Swapcase:", company.swapcase())
```

```python

# 9. Slice out the first word
print("After slicing first word:", company[7:])  # Removes 'Coding '
```

```python
# 10. Check if it contains 'Coding'
print("Contains 'Coding'?", 'Coding' in company)
```

```python

# 11. Replace 'Coding' with 'Python'
print("Replace 'Coding' with 'Python':", company.replace('Coding', 'Python'))
```

```python
# 12. Replace in different sentence
print("Replace in new string:", 'Python for Everyone'.replace('Everyone', 'All'))
```

```python
# 13. Split using space
print("Split by space:", company.split())
```

```python
# 14. Split comma-separated list
tech_companies = "Facebook, Google, Microsoft, Apple, IBM, Oracle, Amazon"
print("Split by comma:", tech_companies.split(', '))
```

```python
# 15. Character at index 0
print("Character at index 0:", company[0])

```

```python
# 16. Last index
print("Character at last index:", company[-1])
```

```python
# 17. Character at index 10
print("Character at index 10:", company[10])
```

```python

# 18. Acronym for 'Python For Everyone'
pfe = ''.join([word[0] for word in 'Python For Everyone'.split()])
print("Acronym for 'Python For Everyone':", pfe)
```

```python
# 19. Acronym for 'Coding For All'
cfa = ''.join([word[0] for word in 'Coding For All'.split()])
print("Acronym for 'Coding For All':", cfa)
```

```python
# 20. Position of first 'C'
print("First occurrence of 'C':", company.index('C'))
```

```python
# 21. Position of first 'F'
print("First occurrence of 'F':", company.index('F'))
```

```python
# 22. Last occurrence of 'l' using rfind
print("Last occurrence of 'l':", 'Coding For All People'.rfind('l'))
```

```python

# 23. First occurrence of 'because'
sentence = 'You cannot end a sentence with because because because is a conjunction'
print("First 'because' index:", sentence.find('because'))
```

```python
# 24. Last occurrence of 'because'
print("Last 'because' index:", sentence.rindex('because'))
```

```python

# 25. Slice out 'because because because'
first = sentence.find('because')
last = first + len('because because because')
print("Sliced phrase:", sentence[first:last])
```

```python
# 26. Check if starts with 'Coding'
print("Starts with 'Coding'?", company.startswith('Coding'))
```

```python
# 27. Check if ends with 'coding'
print("Ends with 'coding'?", company.endswith('coding'))
```

```python
# 28. Remove trailing spaces
company_space = '   Coding For All      '
print("Without spaces:", company_space.strip())
```

```python
# 29. Which isidentifier
print("'30DaysOfPython' isidentifier?", '30DaysOfPython'.isidentifier())
print("'thirty_days_of_python' isidentifier?", 'thirty_days_of_python'.isidentifier())
```

```python
# 30. Join Python libraries with hash and space
libraries = ['Django', 'Flask', 'Bottle', 'Pyramid', 'Falcon']
joined_libs = ' # '.join(libraries)
print("Joined libraries:", joined_libs)
```

```python
# 31. New line escape sequence
print("I am enjoying this challenge.\nI just wonder what is next.")
```

```python

# 32. Tab escape sequence
print("Name\t\tAge\tCountry\t\tCity")
print("Asabeneh\t250\tFinland\t\tHelsinki")
```

```python
# 33. String formatting (circle area)
radius = 10
area = 3.14 * radius ** 2
print("The area of a circle with radius {} is {} meters square.".format(radius, int(area)))

```

```python
# 34. String formatting for math operations
a, b = 8, 6
print("{} + {} = {}".format(a, b, a + b))
print("{} - {} = {}".format(a, b, a - b))
print("{} * {} = {}".format(a, b, a * b))
print("{} / {} = {:.2f}".format(a, b, a / b))
print("{} % {} = {}".format(a, b, a % b))
print("{} // {} = {}".format(a, b, a // b))
print("{} ** {} = {}".format(a, b, a ** b))
```
